Barplot: 막대그래프
- 목적: 범주형 데이터의 값을 막대 높이로 비교
- X축: 범주(카테고리)
- Y축: 수치(비교 값)

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [10, 25, 15, 30]

plt.bar(categories, values)
plt.title('Barplot Example')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

Boxplot: 상자그림
- 목적: 데이터 분포를 요약(중앙값, 사분위수, 이상치 포함)
- 특징: 중앙값, 1/4분위수(Q1), 3/4분위수(Q3), 이상치

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(0, 1, 100)

plt.boxplot(data)
plt.title('Boxplot Example')
plt.show()
```

Violinplot: 바이올린 플롯
- 목적: Boxplot에 데이터의 분포 밀도(KDE)를 추가
- 특징: 중앙값, 사분위수, 분포의 모양(두께) 표시

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('tips')

sns.violinplot(x='day', y='total_bill', data=data)
plt.title('Violinplot Example')
plt.show()
```

Countplot: 카운트 플롯
- 목적: 범주형 데이터의 빈도수(count) 표현
- 특징: x축 또는 y축에 범주별 데이터 개수를 세어 막대로 표시

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('titanic')

sns.countplot(x='class', data=data)
plt.title('Countplot Example')
plt.show()
```

Pie chart: 파이 차트
- 목적: 전체 대비 각 부분의 비율 표현
- 특징: 원 형태를 비율만큼 나누어 색상으로 구분

```python
import matplotlib.pyplot as plt

labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [15, 30, 45, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart Example')
plt.axis('equal')
plt.show()
```

Heatmap: 히트맵
- 목적: 2차원 데이터(행렬)의 값을 색상으로 표현
- 특징: 값이 크고 작을수록 다른 색상으로 나타남

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

data = np.random.rand(5, 5)

sns.heatmap(data, annot=True, cmap='coolwarm')
plt.title('Heatmap Example')
plt.show()
```

Scatterplot: 산점도
- 목적: 두 연속형 변수 간의 관계를 점으로 표현
- 특징: 한 점은 x값과 y값의 쌍을 의미, 상관관계 또는 방향성 파악 가능

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.title('Scatterplot Example')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```
