```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
df = pd.read_csv("..\\data\\AAME.csv")
```

```python
df = df.drop(columns=['Date', 'Adj Close', 'Volume'])
```

```python
df = df[df['Open'] != 0]
df = df.dropna()
df_trimmed = df[["Close"]]
```

```python
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df[['Close']])
```

```python
def make_sample(data, window):
    X, y = [], []
    for i in range(len(data) - window):
        X.append(data[i:i+window])
        y.append(data[i+window])
    return np.array(X), np.array(y)
```

```python
window_size = 10
X, y = make_sample(scaled, window_size)
```

```python
X = X.reshape(-1, window_size, 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False)
```

```python
model = tf.keras.Sequential([
    tf.keras.Input(shape=(window_size, 1)),
    tf.keras.layers.Conv1D(filters=64, kernel_size=3, activation='relu'),
    tf.keras.layers.MaxPooling1D(pool_size=2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(50, activation='relu'),
    tf.keras.layers.Dense(1)
])
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(X_train, y_train, epochs=20, batch_size=16, validation_data=(X_test, y_test))
```

```python
predicted = model.predict(X_test)
predicted_inv = scaler.inverse_transform(predicted)
y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 1))
```

```python
plt.figure(figsize=(10, 5))
plt.plot(y_test_inv, label='Actual')
plt.plot(predicted_inv, label='Predicted')
plt.title('1D CNN Prediction of Close Price')
plt.xlabel('Time Step')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
rmse = np.sqrt(mean_squared_error(y_test_inv, predicted_inv))
mae = mean_absolute_error(y_test_inv, predicted_inv)
r2 = r2_score(y_test_inv, predicted_inv)
print(f"🔹 RMSE: {rmse:.4f}")
print(f"🔹 MAE: {mae:.4f}")
print(f"🔹 R² Score: {r2:.4f}")
```
